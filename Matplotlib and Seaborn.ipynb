# <u>Matplotlib</u>


Matplotlib is Python’s foundational 2D plotting library. It powers or influences nearly all other plotting libraries in Python (``Seaborn``, Pandas .plot, Cartopy, etc.).
Its key goals:

- Provide MATLAB-like plotting in Python

- Full control over every pixel (low-level + high-level APIs)

- Support multiple backends (interactive, notebook, static images)

(see matplotlib Gallery
http://matplotlib.org/ or https://matplotlib.org/stable/tutorials/pyplot.html)




## The Two Main APIs

### <u>Pyplot (state-based)</u>

**Useful for quick work, interactive sessions, notebooks**

##### Single plots
 
```python
# Import module pyplot with alias 
import matplotlib.pyplot as plt

# 1. Prepare data
import numpy as np # numpy often used for data
import pandas as pd # data also often comes as dataframe
x = [...] 
y = [...]
z = [...]

# 2. Figure size
plt.figure(figsize=(width, height))

# 3. Some Basic Plots -----------------------------------------------------------

# 📈 Line plot
# Data: numeric x and y of same length
plt.plot(
    x, y,
    color=None,
    linestyle='-',
    linewidth=1.5,
    marker=None,
    label=None
)

# 🔵 Scatter plot
# Data: numeric x and y of same length
plt.scatter(
    x, y,
    c=None,          # color or array for colormap
    s=None,          # marker size
    alpha=None,      # transparency
    marker=None,
    label=None
)

# 📊 Bar plot (vertical)
# Data: categorical or numeric x + numeric heights
plt.bar(
    x,
    height=y,
    width=0.8,
    color=None,
    align='center',
    label=None
)

# 📊 Bar plot (horizontal)
# Data: categorical or numeric y + numeric widths
plt.barh(
    y=x, # categories
    width=y, # bar lengths
    height=0.8,
    color=None,
    label=None
)

# 🌱 Stem plot
# Data: numeric discrete x and y
plt.stem(
    x, y,
    linefmt=None,
    markerfmt=None,
    basefmt=None,
    label=None
)

# 🌊 Fill between curves
# Data: numeric x and two numeric curves y1, y2
plt.fill_between(
    x, y, z,
    where=None,
    color=None,
    alpha=None,
    label=None
)

# 🗻 Stack plot
# Data: x numeric + multiple y sequences
plt.stackplot(
    x, y,
    colors=None,
    labels=None,
    alpha=None
)

# 🪜 Stairs (step function)
# Data: numeric edges and values
plt.stairs(
    values=y,
    edges=x,
    orientation='vertical',
    label=None
)

# 📉 Histogram
# Data: 1D numeric array
plt.hist(
    x,
    bins=10,
    range=None,
    density=False,
    color=None,
    alpha=None,
    label=None
)

# 📦 Boxplot
# Data: 1D or 2D numeric data
plt.boxplot(
    x,
    vert=True,
    patch_artist=False,
    showmeans=False,
    labels=None
)

# 📏 Error bar plot
# Data: numeric x, y and error values
plt.errorbar(
    x, y,
    yerr=z,
    xerr=None,
    fmt='o',
    ecolor=None,
    capsize=None,
    label=None
)

# 🎻 Violin plot
# Data: 1D or grouped numeric distributions
plt.violinplot(
    x,
    showmeans=False,
    showmedians=False,
    showextrema=True
)

# 🥧 Pie chart
# Data: positive numeric sizes
plt.pie(
    y,
    labels=None,
    autopct=None,
    startangle=0,
    explode=None
)

# 📈 ECDF (Matplotlib ≥ 3.8)
# Data: 1D numeric array
plt.ecdf(
    x,
    complementary=False,
    label=None
)

# --------------------------------------------------------------------------------

# 4. Set options
plt.xlabel("X-axis") # set x axis label
plt.ylabel("Y-axis") # set y axis label
plt.title("My Plot") # set figure title
plt.grid(True) # show grid or not
plt.tick_params(axis="both", which="major", labelsize=9) # Ticks styling
plt.legend(loc="best") # Add legend (only if labels were provided in basic plots)
# loc=1 means upper right corner
# loc=2 means upper left corner
# loc=3 means lower left corner
# loc=4 means lower right corner


plt.tight_layout() # Layout tightening
plt.xlim(start,stop) # set x axis limit
plt.ylim(start,stop) # set y axis limit
plt.savefig("filename.png",dpi=300,bbox_inches="tight",facecolor="white") # save figure inside a new file

# 5. Show plot
plt.show()

```

##### Multiple plots

```python
# Import module pyplot with alias 
import matplotlib.pyplot as plt

# 1. Prepare data
import numpy as np
import pandas as pd

x = [...] 
y = [...]

# 2. Create figure and axes grid
fig, axes = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(width, height),
    sharex=False,
    sharey=False
)


# axes is a 2D array when nrows > 1 and ncols > 1
# Access examples:
# axes[0,0], axes[0,1], axes[1,0], axes[1,1]
# or
# if fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(height, width))
# then we have axes[0], axes[1]


# alternative to fig, axes = plt.subplots(nrows, ncols, figsize=(height, width)) is 
# fig = plt.figure() to create empty canvas
# axes = fig.add_axes([left,bottom,width,height])

# 3. Basic plots on individual axes --------------------------------------------

# 📈 Line plot
axes[0, 0].plot(x, y, label="Line")

# 🔵 Scatter plot
axes[0, 1].scatter(x, y, label="Scatter")

# 📊 Bar plot
axes[1, 0].bar(x, y, label="Bar")

# 📉 Histogram
axes[1, 1].hist(x, bins=10, label="Hist")

# -----------------------------------------------------------------------------

# 4. Set options per subplot
number = 1
for ax in axes.flat:
    ax.set_title(f"Subplot number {number}")
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.grid(True)
    ax.tick_params(axis="both", which="major", labelsize=9)
    ax.legend(loc="best") # only if plots are labeled
    number += 1

# 5. Global figure options
fig.suptitle("Multiple Plots Example")
plt.tight_layout()

# Optional save
fig.savefig("filename.png", dpi=300, bbox_inches="tight", facecolor="white")

# 6. Show plot
plt.show()

```

````python
#help(plt.plot)
import matplotlib.pyplot as plt
import numpy as np

# Pyplot (state-based)

# Prep data
x = np.linspace(-10,10,100)
y = np.sin(x)

# Adjust figure size
width, height = 5,4
plt.figure(figsize=(width,height))

# Basic line plot
plt.plot(x,y,color="red",linestyle="-", linewidth=1.5, marker="+",label="Sinus curve")

# Set options
plt.xlabel("x")
plt.ylabel("y")
plt.title("My plot")
plt.grid(True)
plt.legend(loc=2)
plt.tight_layout()
plt.xlim(0,10)

plt.show()


````

````python
# Pyplot (state-based)

# Prep data
x=np.linspace(0,10,100)
y1=x**2 - 2*x
y2 = x**2

plt.plot(x, y1, c="r", label="x^2 - 2x")
plt.plot(x, y2, c="g", label="x^2")

plt.title("Multiple lines")
plt.xlabel("x")

plt.legend(loc=4)
# loc=1 means upper right corner
# loc=2 means upper left corner
# loc=3 means lower left corner
# loc=4 means lower right corner

plt.show()
````

````python
# OOP API

# Prep data
x = np.random.normal(0, 5, 1000)

# Figure size
fig, ax = plt.subplots(figsize=(7, 4))

# Basic plot
ax.hist(
    x,
    bins=10,
    density=False,
    color="yellow",
    alpha=0.5, # color intensity
    label="Hist"
)

ax.set_xlabel("X-axis")
ax.set_ylabel("Frequency")
ax.set_title("My Plot")
ax.legend(loc="best")

plt.show()

````

````python
# OOP API

# Prep data
# np.random.normal(loc, scale, size) with loc=(3,4,5), scale=(1.25, 1.00, 1.25) and size=(100,3)=(rows,columns)
x = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3)) 

# Figure size
fig, ax = plt.subplots(figsize=(7, 4))

# Basic plot
ax.boxplot(
    x,
    positions=[1,3,5],
    widths=1.5,
    label="Boxplot"
)

ax.set_xlabel("Positions")
ax.set_ylabel("x")
ax.set_title("Boxplot")
ax.legend(loc="best")

plt.show()
````

````python
# OOP API

# Prep data
x = np.linspace(-10,10,50)
y = x**2

# Define figure size and subaxis
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7, 6))  # 1 row, 2 columns

# Subplots
axes[0].plot(x,y, "g*--")
axes[0].set_title("Left Plot")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

axes[1].plot(x,y+2*x, c="r")
axes[1].set_title("Right Plot")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

fig.suptitle("Multiple Plots Example")
plt.tight_layout()

plt.show()
````
