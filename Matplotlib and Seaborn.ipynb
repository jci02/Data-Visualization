# <u>Matplotlib</u>


Matplotlib is Python’s foundational 2D plotting library. It powers or influences nearly all other plotting libraries in Python (``Seaborn``, Pandas .plot, Cartopy, etc.).
Its key goals:

- Provide MATLAB-like plotting in Python

- Full control over every pixel (low-level + high-level APIs)

- Support multiple backends (interactive, notebook, static images)

(see matplotlib Gallery
http://matplotlib.org/ or https://matplotlib.org/stable/tutorials/pyplot.html)




## The Two Main APIs

### <u>Pyplot (state-based)</u>

**Useful for quick work, interactive sessions, notebooks**

##### Single plots
 
```python
# Import module pyplot with alias 
import matplotlib.pyplot as plt

# 1. Prepare data
import numpy as np # numpy often used for data
import pandas as pd # data also often comes as dataframe
x = [...] 
y = [...]
z = [...]

# 2. Figure size
plt.figure(figsize=(width, height))

# 3. Some Basic Plots -----------------------------------------------------------

# 📈 Line plot
# Data: numeric x and y of same length
plt.plot(
    x, y,
    color=None,
    linestyle='-',
    linewidth=1.5,
    marker=None,
    label=None
)

# 🔵 Scatter plot
# Data: numeric x and y of same length
plt.scatter(
    x, y,
    c=None,          # color or array for colormap
    s=None,          # marker size
    alpha=None,      # transparency
    marker=None,
    label=None
)

# 📊 Bar plot (vertical)
# Data: categorical or numeric x + numeric heights
plt.bar(
    x,
    height=y,
    width=0.8,
    color=None,
    align='center',
    label=None
)

# 📊 Bar plot (horizontal)
# Data: categorical or numeric y + numeric widths
plt.barh(
    y=x, # categories
    width=y, # bar lengths
    height=0.8,
    color=None,
    label=None
)

# 🌱 Stem plot
# Data: numeric discrete x and y
plt.stem(
    x, y,
    linefmt=None,
    markerfmt=None,
    basefmt=None,
    label=None
)

# 🌊 Fill between curves
# Data: numeric x and two numeric curves y1, y2
plt.fill_between(
    x, y, z,
    where=None,
    color=None,
    alpha=None,
    label=None
)

# 🗻 Stack plot
# Data: x numeric + multiple y sequences
plt.stackplot(
    x, y,
    colors=None,
    labels=None,
    alpha=None
)

# 🪜 Stairs (step function)
# Data: numeric edges and values
plt.stairs(
    values=y,
    edges=x,
    orientation='vertical',
    label=None
)

# 📉 Histogram
# Data: 1D numeric array
plt.hist(
    x,
    bins=10,
    range=None,
    density=False,
    color=None,
    alpha=None,
    label=None
)

# 📦 Boxplot
# Data: 1D or 2D numeric data
plt.boxplot(
    x,
    vert=True,
    patch_artist=False,
    showmeans=False,
    labels=None
)

# 📏 Error bar plot
# Data: numeric x, y and error values
plt.errorbar(
    x, y,
    yerr=z,
    xerr=None,
    fmt='o',
    ecolor=None,
    capsize=None,
    label=None
)

# 🎻 Violin plot
# Data: 1D or grouped numeric distributions
plt.violinplot(
    x,
    showmeans=False,
    showmedians=False,
    showextrema=True
)

# 🥧 Pie chart
# Data: positive numeric sizes
plt.pie(
    y,
    labels=None,
    autopct=None,
    startangle=0,
    explode=None
)

# 📈 ECDF (Matplotlib ≥ 3.8)
# Data: 1D numeric array
plt.ecdf(
    x,
    complementary=False,
    label=None
)

# --------------------------------------------------------------------------------

# 4. Set options
plt.xlabel("X-axis") # set x axis label
plt.ylabel("Y-axis") # set y axis label
plt.title("My Plot") # set figure title
plt.grid(True) # show grid or not
plt.tick_params(axis="both", which="major", labelsize=9) # Ticks styling
plt.legend(loc="best") # Add legend (only if labels were provided in basic plots)
# loc=1 means upper right corner
# loc=2 means upper left corner
# loc=3 means lower left corner
# loc=4 means lower right corner


plt.tight_layout() # Layout tightening
plt.xlim(start,stop) # set x axis limit
plt.ylim(start,stop) # set y axis limit
plt.savefig("filename.png",dpi=300,bbox_inches="tight",facecolor="white") # save figure inside a new file

# 5. Show plot
plt.show()

```

##### Multiple plots

```python
# Import module pyplot with alias 
import matplotlib.pyplot as plt

# 1. Prepare data
import numpy as np
import pandas as pd

x = [...] 
y = [...]

# 2. Create figure and axes grid
fig, axes = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(width, height),
    sharex=False,
    sharey=False
)


# axes is a 2D array when nrows > 1 and ncols > 1
# Access examples:
# axes[0,0], axes[0,1], axes[1,0], axes[1,1]
# or
# if fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(height, width))
# then we have axes[0], axes[1]


# alternative to fig, axes = plt.subplots(nrows, ncols, figsize=(height, width)) is 
# fig = plt.figure() to create empty canvas
# axes = fig.add_axes([left,bottom,width,height])

# 3. Basic plots on individual axes --------------------------------------------

# 📈 Line plot
axes[0, 0].plot(x, y, label="Line")

# 🔵 Scatter plot
axes[0, 1].scatter(x, y, label="Scatter")

# 📊 Bar plot
axes[1, 0].bar(x, y, label="Bar")

# 📉 Histogram
axes[1, 1].hist(x, bins=10, label="Hist")

# -----------------------------------------------------------------------------

# 4. Set options per subplot
number = 1
for ax in axes.flat:
    ax.set_title(f"Subplot number {number}")
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.grid(True)
    ax.tick_params(axis="both", which="major", labelsize=9)
    ax.legend(loc="best") # only if plots are labeled
    number += 1

# 5. Global figure options
fig.suptitle("Multiple Plots Example")
plt.tight_layout()

# Optional save
fig.savefig("filename.png", dpi=300, bbox_inches="tight", facecolor="white")

# 6. Show plot
plt.show()

```

# <u>Seaborn</u>

Seaborn is a high-level statistical data visualization library built on top of Matplotlib.

Its goals:

- Automatic statistical computations

- Provide beautiful default styles

- Work seamlessly with pandas DataFrames

- Provides advanced categorical & regression plots

- Add plot types Matplotlib doesn’t have by default

Think of Seaborn as:

**matplotlib + statistics + styling + intuitive dataset handling**

## 1.Seaborn Architecture

Seaborn is organized into specialized modules:

| Module                  | Purpose                                    |
| ----------------------- | ------------------------------------------ |
| **relplot()**           | Relationships: scatter, line               |
| **displot()**           | Distributions: hist, KDE                   |
| **catplot()**           | Categorical data                           |
| **regplot(), lmplot()** | Regression & trend fitting                 |
| **jointplot()**         | Matched bivariate + marginal distributions |
| **pairplot()**          | Scatter matrix across several variables    |
| **clustermap()**        | Heatmap + hierarchical clustering          |
| **heatmap()**           | Standard heatmap                           |
| **FacetGrid()**         | Multi-panel plotting                       |
| **PairGrid()**          | Grid of subplot types                      |
| **JointGrid()**         | Custom bivariate + marginal plots          |

All of them sit on top of Matplotlib, so you can always modify axes after the fact


## 2.Seaborn + Pandas Workflow

**Seaborn expects data in DataFrames**

Typical Seaborn call:
```python
import seaborn as sns
sns.scatterplot(data=df, x="height", y="weight", hue="gender")
sns.__version__ # see version 

```
This is much simpler than passing arrays because:

- It uses column names, not raw arrays

- It automatically maps columns to visual properties

## 3.Themes and Aesthetics (Plot Styles)

Seaborn gives you full style control:

<u>Built-in themes</u>

- ``darkgrid (default)``

- ``whitegrid``

- ``dark``

- ``white``

- ``ticks``

Apply globally:

```python
sns.set_theme(style="whitegrid", context="notebook")
```

<u>Contexts</u>

Controls scale:

- ``paper``

- ``notebook (default)``

- ``talk``

- ``poster``

For larger fonts:
```python
sns.set_context("talk")

```

<u>Color palettes</u>

Carefully designed palettes:

- Sequential (``Blues``, ``Greens``, ``viridis``)

- Diverging (``coolwarm``, ``bwr``, ``RdBu``)

- Categorical (``Set1``, ``tab10``, ``Paired``)

- Special palettes: ``color_palette("husl")``, ``cubehelix_palette()``, ``light_palette()``, ``dark_palette()``





## 4. Core Plot Families in Seaborn

**Below are all the major plot types**

### 4.1 Relational Plots (Relationships)

``scatterplot()``

- Better scatter with categories/hue/size/style:

```python
sns.scatterplot(data=df, x="A", y="B", hue="group", size="value")
```


``lineplot()``
- Statistical aggregation automatically:

```python
sns.lineplot(data=df, x="time", y="score", ci=95)

```

``replot()``
- Figure-level interface that can produce either scatter or line:

```python
sns.relplot(data=df, x="A", y="B", hue="group", col="category")

```
---

### 4.2 Distribution Plots

``histplot()``

- Modern replacement for distplot.

``kdeplot()``

- Kernel Density Estimate (smooth distribution curve).


``ecdfplot()``  
- Empirical cumulative distribution curve.

``rugplot()``  
- Small tick marks showing individual observations.

``displot()``  
- Figure-level interface for histograms, KDE, and ECDF.

---

### 4.3 Categorical Plots

``boxplot()``  
- Classic box-and-whisker summary.

``violinplot()``  
- KDE-based symmetrical distribution plot.

``stripplot()``  
- Raw data points (optionally with jitter).

``swarmplot()``  
- Non-overlapping scatter for categorical variables.

``barplot()``  
- Mean + confidence interval over categories.

``countplot()``  
- Counts of each category level.

``pointplot()``  
- Point estimates + confidence intervals across categories.

``lineplot()``  
- Line plot with statistical aggregation.

``catplot()``  
- Figure-level interface for all categorical plots.

---

### 4.4 Regression and Statistical Modeling

``regplot()``  
- Scatterplot + optional regression line.

``lmplot()``  
- Figure-level version of regplot using FacetGrid.

``residplot()``  
- Visualizes residuals of a regression model.

``regplot(order=...)``  
- Polynomial or robust regression fits.

---

### 4.5 Multivariate & Grid Plots

``pairplot()``  
- Scatterplot matrix for multivariate data.

``jointplot()``  
- Joint distribution with marginal hist/KDE.

``PairGrid``  
- Fully customizable pairplot grid.

``FacetGrid``  
- Multi-plot grid based on row/column categories.

---

### 4.6 Heatmaps and Matrices

``heatmap()``  
- Visualizes matrices (e.g., correlations).

``clustermap()``  
- Heatmap with hierarchical clustering.




## 5. Customization

**Since Seaborn is built on Matplotlib, you can modify anything:**

```python
ax = sns.scatterplot(data=df, x="A", y="B")
ax.set_title("My Plot")

```

You can also adjust:

- Figure size

- Legends

- Color palettes

- Axis labels

- Axis limits

- Annotations

- Patch objects (bars, boxes, etc.)

## 6. Statistical Features

**Seaborn does a lot automatically:**

- Compute confidence intervals (ci parameter)

- Bootstrapping for categorical plots

- Automatic aggregation with estimator=mean

- Smoothing (regplot = confidence band)

- KDE estimation (kdeplot)

- Clustering (clustermap)

## 7. Color Mapping (hue, style, size)

**Seaborn allows multiple variables encoded visually:**

```python
sns.scatterplot(
    data=df,
    x="A",
    y="B",
    hue="category",
    style="type",
    size="value"
)

```
Seaborn automatically handles:

- color palettes

- marker types

- scaling

````python
# Figure size
#plt.figure(figsize=(14,7))

# Create a histogram using Seaborn
sns.histplot(tips['tip'], kde=False, color="b")  # kde=True adds a density curve

# Customize the plot
plt.xlabel('Tip Amount (in Dollars)')
plt.ylabel('Frequency')
plt.title('Distribution of Tips by customers')

plt.tight_layout()

# Show the plot
plt.show()
````

### jointplot

jointplot() lets you combine two displots for two variables. You can choose what kind of plot to use for the comparison, such as: 
* "scatter" 
* "reg"
* "resid" 
* "kde" 
* "hex"

````python

sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')  # we visualize x and y using jointplot to see the realation betwen x,y
plt.show()


sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') # relation between x=total_bill and y=tip
# the total_bill is concentrated again between 10 and 20 and the tip is concentrated between 2 and 4
plt.show()


sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # reg is a regression line
plt.show()
````

### pairplot

pairplot can show how two variables relate to each other in a whole dataframe (only for the numeric columns) and can use a color hue parameter (for the columns with categories). 

````python
sns.pairplot(tips) # the relation between the variable itself on the main diagonal nad the scatter plot the relation between
                   # two different variable
plt.show()


sns.pairplot(tips,hue='sex',palette='coolwarm') # pairplot accepts another attribut "hue"
plt.show()
````
