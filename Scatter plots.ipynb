```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
insurance_data=pd.read_csv("insurance.csv")
insurance_data
```

```python
#help(sns.scatterplot)
sns.scatterplot(data=insurance_data,x="bmi",y="charges",markers="+",hue="smoker",palette="viridis")
plt.title("Relation between Body-mass-index and charges")
plt.show()


#help(sns.regplot)
sns.regplot(data=insurance_data,x="bmi",y="charges")
plt.title("Relation between Body-mass-index and charges")
plt.show()
```

```python
#help(sns.lmplot)
sns.lmplot(data=insurance_data,x="bmi",y="charges",hue="smoker")
plt.title("Relation between Body-mass-index and charges")
plt.show()
```

```python
#help(sns.stripplot)
#help(sns.swarmplot)
sns.stripplot(data=insurance_data,x="smoker",y="charges",color="orange")
plt.title("Relation between Body-mass-index and charges")
plt.show()
```

```python
candy_data = pd.read_csv("candy.csv")
candy_data
```

```python
candy_data = pd.read_csv("candy.csv",index_col=0)
candy_data
```

```python

cmp = candy_data.winpercent[candy_data["competitorname"].isin(["3 Musketeers", "Almond Joy"])]
if (cmp.loc[1] - cmp.loc[3]) > 0: 
    print(f"Which candy was more popular with survey responders \"3 Musketters\" or \"Almond Joy\"?: 3 Musketeers")
else:
    print(f"Which candy was more popular with survey responders \"3 Musketters\" or \"Almond Joy\"?: Almond Joy")

# Which candy has higher sugar content?: "Air Heads" or "Baby Ruth"
cmp = candy_data.sugarpercent[candy_data["competitorname"].isin(["Air Heads", "Baby Ruth"])]
if (cmp.iloc[0] - cmp.iloc[1]) > 0: 
    print(f"Which candy has more sugar \"Air Heads\" or \"Baby Ruth\"?: Air Heads")
else:
    print(f"Which candy was more popular with survey responders \"Air Heads\" or \"Baby Ruth\"?: Baby Ruth")
```

```python
sns.scatterplot(data=candy_data,x="sugarpercent",y="winpercent",color="grey")
plt.show()

sns.lmplot(data=candy_data,x="sugarpercent",y="winpercent")
plt.show()
```

```python
sns.scatterplot(data=candy_data,x="pricepercent",y="winpercent",hue="chocolate",color="grey")
plt.show()

sns.lmplot(data=candy_data,x="pricepercent",y="winpercent",hue="chocolate")
plt.show()
```

```python
sns.swarmplot(data=candy_data,x="chocolate",y="winpercent",color="orange")
plt.show()
```
