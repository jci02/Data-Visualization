# <u>Bar Charts</u>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
flight_data = pd.read_csv("flight_delays.csv",index_col ="Month")
flight_data
```

```python
#help(sns.barplot)
```

```python
# Set size
plt.figure(figsize=(9,7))

# Barplot
sns.barplot(x=flight_data.index,y=flight_data["NK"],color="grey")

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flight per Month")

plt.ylabel("Arrival delay in minutes")

plt.show()
```

```python
#help(sns.heatmap)
# Set size
plt.figure(figsize=(14,7))

# Barplot
sns.heatmap(data=flight_data,annot=True)

# Add title
plt.title("Average Arrival Delay for each Airline by Month")

plt.xlabel("Airline")

plt.show()
```

```python
ign_data = pd.read_csv("ign_scores.csv",index_col="Platform")
ign_data
```

```python
print(f"Highest average score received by PC games for any genre?: {ign_data.loc["PC"].max()} by {ign_data.loc["PC"].idxmax()}")
print(f"On PlayStatio Vita platform, which genre has the lowest average score?: {ign_data.loc["PlayStation Vita"].idxmin()}")

#help(sns.barplot)

plt.figure(figsize=(8,5))

sns.barplot(data=ign_data,x="Racing",y=ign_data.index,hue = ign_data.index,palette="coolwarm")

plt.title("Average score for racing games for each platform")
plt.show()


plt.figure(figsize=(8,5))

sns.barplot(data=ign_data,x="Racing",y=ign_data.index,hue = ign_data.index,palette="viridis")

plt.title("Average score for racing games for each platform")
plt.show()


plt.figure(figsize=(8,5))

sns.barplot(data=ign_data,x="Racing",y=ign_data.index,hue = ign_data.index,palette="tab10")

plt.title("Average score for racing games for each platform")
plt.show()



plt.figure(figsize=(8,5))

sns.barplot(data=ign_data,x="Racing",y=ign_data.index,hue = ign_data.index,palette="Set2")

plt.title("Average score for racing games for each platform")
plt.show()
```

```python
#help(sns.heatmap)

plt.figure(figsize=(10,10))

sns.heatmap(data=ign_data,annot=True)

plt.title("Average scores by genre")

plt.xlabel("Genre")

# Rotate ticks
#plt.xticks(rotation=) # x-axis labels angle
plt.yticks(rotation=0) # y-axis labels angle

plt.tight_layout()

plt.show()
```
