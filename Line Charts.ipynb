```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
spotify_data = pd.read_csv("spotify.csv")
spotify_data.info() # column Date is of type object
```

```python
spotify_data = pd.read_csv("spotify.csv",parse_dates=["Date"])
spotify_data.info() # column Date is of type datetime64[ns] now
spotify_data
```

```python
spotify_data = pd.read_csv("spotify.csv",parse_dates=["Date"],index_col="Date")
spotify_data
```

```python
print(list(spotify_data.columns))
spotify_data.isna().sum()
```

```python
# Plot the data using index "Date" as x-axis (there are many NaNs in the beginning of 2017)
plt.figure(figsize=(12,7))
sns.lineplot(data=spotify_data)
plt.title("Daily Global Streams of Songs from 2017 to 2018")
plt.ylabel("Daily Streams")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(data=spotify_data["Shape of You"], label="Shape of you")
sns.lineplot(data=spotify_data["Despacito"], label="Despacito")
plt.title("Daily Global Streams of Songs from 2017 to 2018")
plt.show()
```

```python
museum_data = pd.read_csv("museum_visitors.csv",index_col="Date",parse_dates=True)
print(museum_data.index.dtype) # turns from object to datetime64[ns] with parse_dates=True
museum_data.head()
```

```python
print(f"Number of visitors of Chinese American Museum for July 2018: {museum_data.loc["2018-07-01","Chinese American Museum"]}")
print(f"In October 2018 how many more visitors did Avila Adobe receive than the Firehouse Museum: {museum_data.loc["2018-10-01","Avila Adobe"]-museum_data.loc["2018-10-01","Firehouse Museum"]}")
museum_data.tail()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(data=museum_data)
plt.title("Visitorcount over the years")
plt.ylabel("Visitors")
plt.show()

plt.figure(figsize=(6,4))
sns.lineplot(data=museum_data["Avila Adobe"],color="orange")
plt.title("Visitorcount over the years of Avila Adobe")
plt.ylabel("Visitors to Avial Adobe")
plt.show()
```
